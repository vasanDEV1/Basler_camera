```python
img=cv2.imread("C://Users//krvas//Desktop//tst_image_resize.png")
#plt.imshow(img)
img_grey=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img_grey)
```

```python
_,thresh1 = cv2.threshold(img_grey, 150, 255, cv2.THRESH_BINARY)
contours, hierarchy = cv2.findContours(thresh1, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
con_img=cv2.drawContours(img_grey, contours, -1, (0,0,255), 4)
plt.imshow(con_img)
```

```python

for contour in contours:
    area = cv2.contourArea(contour)
    if 100000 < area < 200000:
        #print(contour)  
        con = np.array(contour)
        points = con.reshape(-1, 2)
        points_list = points.tolist()
        #print(points_list)
centre = [x,y]
rad = [math.dist(point,centre) for point in points_list]
rad
```

```python

# atan2 function takes postive x-axis as reference axis and center[x,y] as reference point and calculate the angle of contour points wrt centre
angles = [math.atan2(point[1] - centre[1], point[0] - centre[0]) for point in points_list]
angles = np.array(angles)  # Convert to NumPy array for easier manipulation

#Sort angles and corresponding radii
sorted_indices = np.argsort(angles)
sorted_angles = angles[sorted_indices]
sorted_radii = np.array(rad)[sorted_indices]
sorted_angles = np.degrees(sorted_angles)
sorted_radii = sorted_radii * 0.09
print(sorted_angles)

# Plot the variation of radius with respect to angle
plt.figure(figsize=(8, 6))
plt.plot(sorted_angles, sorted_radii, label="Radius vs Angle")
plt.xlabel("Angle")
plt.ylabel("Radius")
plt.title("Variation of Radius with Respect to Angle")
plt.legend()
plt.grid()
plt.show()
```

```python
#taking 20mm as geometric radius and subracting from the calculated radius
new_sorted_radii = [ i-20 for i in sorted_radii]
plt.figure(figsize=(8,6))
plt.plot(sorted_angles,new_sorted_radii)
plt.xlabel("Angle")
plt.ylabel("err_in_rad_wrt_geo_rad") 
plt.grid()
plt.show()
```
